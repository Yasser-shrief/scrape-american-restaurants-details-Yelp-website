```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
```

```python

#create empty dict to save our data in CSV file
our_data={}
film_no=0
page_no=0
url='https://www.yelp.com/search?find_desc=American%20(Traditional)&find_loc=New%20York%2C%20NY&start=0'
while True:
    response=requests.get(url)
    response
    data=response.text
    soup=BeautifulSoup(data,'html.parser')
    #i want to scrape the restaurant's name, phone number, and email address 
    resturants=soup.find_all('ul',attrs={'class':'lemon--ul__373c0__1_cxs undefined list__373c0__2G8oH'})
    for resturant in resturants:                    
        for l in resturant.find_all('div',{'class':'lemon--div__373c0__1mboc arrange__373c0__2C9bH border-color--default__373c0__3-ifU'}):
            for link in l.find_all('a',{'class':'lemon--a__373c0__IEZFH link__373c0__1G70M link-color--inherit__373c0__3dzpk link-size--inherit__373c0__1VFlE'}):
                name=link.text
            for link in l.find_all('div',{'class':'lemon--div__373c0__1mboc scrollablePhotos__373c0__1LEvd arrange__373c0__2C9bH border-color--default__373c0__3-ifU'}):
                for n in link.find_all('p',{'class':'lemon--p__373c0__3Qnnj text__373c0__2Kxyz text-color--black-extra-light__373c0__2OyzO text-align--right__373c0__1f0KI text-size--small__373c0__3NVWO'}):
                    number=n.text
                    if number.startswith("("):
                        tele=number
                  
                #collect all data inside the dict,which the key is job_no OR title
                our_data[film_no]=[name,tele]
                film_no+=1
                print(film_no)
                print('Name:',name,'\nnumber:',tele,'\n\n_________')

    movies=pd.DataFrame.from_dict(our_data , orient='index',columns=['name','tele'])
    #Save it in CSV file
    movies.to_csv('resturant.csv',encoding='utf-8')
    
    #to get the url for next page and replay all the previos steps on it
    page_no+=30
    url='https://www.yelp.com/search?find_desc=American%20(Traditional)&find_loc=New%20York%2C%20NY&start='+str(page_no)
        
    print(url)
    if page_no==4560:
        break

#Note : the total pages shown in website is a 153 pages ,But the reality is that the number of pages is only a 33 page, to this day
    
```
